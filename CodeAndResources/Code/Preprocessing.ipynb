```python
#Observe the features df Zip Code column.
# 5-4 zip code Example 111112222 -> Length 9
# 4-4 zip code 11112222 (Really 011112222) -> Length 8
# 5dash4 zip code 11111-2222 -> Length 10
# 4 zip code 1111 (Really 01111) -> Length 4

for idx , row in features.iterrows():
    zip_code = features.loc[idx,"Zip Code"]
    lz = len(zip_code)

    if lz == 9:
        features.loc[idx,"Zip Code"] = zip_code[0:5]
    elif lz == 10:
        features.loc[idx,"Zip Code"] = zip_code[0:5]
    elif lz == 8:
        features.loc[idx,"Zip Code"] = "0" + zip_code[0:4]
    elif lz == 4:
        features.loc[idx,"Zip Code"] = "0" + zip_code[0:4]

features
```

```python
#Make bins for Surface Parking Spaces
#Based on the array output bins will be Small 0-49 , Large 50-99 , XLarge 100-249 , Jumbo 250-5000
bins = [0,49,99,249,5000]
group_names = ["Small 0-49","Large 50-99","XLarge 100-249", "Jumbo 250-5000"]
features["Surface Sizes"] = pd.cut(features["Surface Parking Spaces"], bins , labels = group_names , include_lowest = True)
features
```

```python
#Create a bin column for the zip codes
zipbins = [0,9999,19999,29999,39999,49999,59999,69999,79999,89999,100000]
group_zipnames = ["Region 0","Region 1","Region 2","Region 3", "Region 4", "Region 5", "Region 6","Region 7","Region 8","Region 9"]
features = features.astype({"Zip Code" : 'float64'})
features["Regions"] = pd.cut(features["Zip Code"], zipbins , labels = group_zipnames , include_lowest = True)
features = features.astype({'Zip Code': 'str'})
features
```

```python
#Create function that takes format DD-MON-YY to YYYY-MM-DD

def to_form(date):
    datelist = date.split("-")
    year = ""
    month = ""
    day = ""
    
    if len(datelist[0]) == 1:
        day = "0" + datelist[0]
    elif len(datelist[0]) == 2:
        day = datelist[0]

    if datelist[1] == "Jan":
        month = "01"
    elif datelist[1] == "Feb":
        month = "02"
    elif datelist[1] == "Mar":
        month = "03"
    elif datelist[1] == "Apr":
        month = "04"
    elif datelist[1] == "May":
        month = "05"
    elif datelist[1] == "Jun":
        month = "06"
    elif datelist[1] == "Jul":
        month = "07"
    elif datelist[1] == "Aug":
        month = "08"
    elif datelist[1] == "Sep":
        month = "09"
    elif datelist[1] == "Oct":
        month = "10"
    elif datelist[1] == "Nov":
        month = "11"
    elif datelist[1] == "Dec":
        month = "12"

    year = "20" + datelist[2]

    return year + "-" + month + "-" + day
```

```python
#Use function to transform dates

for idx, row in features.iterrows():

    features.loc[idx,"End Date"] = to_form(features.loc[idx,"End Date"])
    features.loc[idx,"Start Date"] = to_form(features.loc[idx,"Start Date"])
```

```python
#Convert End Date and Start Date to dt
features['End Date'] = pd.to_datetime(features['End Date'])
features['Start Date'] = pd.to_datetime(features['Start Date'])
features['Length of Lease'] = (features['End Date'] - features['Start Date']).dt.days
```

```python
#Describe Length of Lease
features["Length of Lease"].describe()
```

```python
#Clean for negative values in Length of Lease
features = features.loc[features["Length of Lease"] > 0,:]
features
```

```python
#Create bins for Length of Lease
lease_bins = [0,999,2999,4999,15000]
group_leasenames = ["Short 0-999","Medium 1000-2999","Long 3000-4999","Very Long 5000-15000"]

features["Lease Sizes"] = pd.cut(features["Length of Lease"] , lease_bins , labels = group_leasenames , include_lowest = True)
```

```python
#Create category for top 4 states or other

topstates = ["CA","IL","TX","NY"]
features["State Category"] = ""

for idx, row in features.iterrows():

    if features.loc[idx,"State"] in topstates:
        features.loc[idx,"State Category"] = features.loc[idx,"State"]
    else:
        features.loc[idx,"State Category"] = "Other"

```
